#### 1b. Preliminary run

```python
# reading airliner dataset
df = pd.read_csv('../data/airlinerdata.txt', skipinitialspace=True, header=None)
df = (df.assign(y = df[0].apply(lambda x: x.split(' ')[0])).assign(N = df[0].apply(lambda x: x.split(' ')[3]))).drop([0],axis=1).drop([0],axis=0)
```

```python
# setting up initial data
data = {'y': df['y'].values, 'N': df['N'].values}
```

```python
# setting up hyperparameters initial values for 3 chains
inits = [{'alpha': 0.5, 'beta': 1.5},
         {'alpha': 5, 'beta': 15},
         {'alpha': 10, 'beta': 0.15}]
```

```python
## jags model string equivalent to .bug file
jags_model_string = '''
model {
  for (i in 1:length(y)) {
    lambda[i] ~ dgamma(alpha, beta)
    y[i] ~ dpois(N[i] * lambda[i])
  }
  alpha ~ dgamma(0.001,0.001)
  beta ~ dgamma(0.001,0.001)
}
'''
```

```python
# jags model initialization using pyjags
jags_model \
    = pj.Model(code=jags_model_string,
               init=inits,
               data=data,
               chains=3,
               adapt=0,
               )
```

```python
samples = jags_model.sample(iterations=2000)
```

```python
idata = az.from_pyjags(samples)
```

##### Report the plots you used to monitor and confirm convergence
* Trace plots
* Desity plots
* Gelman plots
* Autocorrelation plots


```python
fig, ax = plt.subplots(3, 2, figsize=(12,5))
az.plot_trace(idata, var_names=['alpha', 'beta', 'lambda'], axes=ax)
plt.tight_layout()
ax[0][1].legend()
plt.show()
```

```python
# Autocorrelation plots for chain # 1
az.plot_autocorr(idata.posterior.sel(chain=1), var_names=['alpha', 'beta', 'lambda'])
plt.tight_layout()
plt.show()
```

```python
az.plot_posterior(idata, var_names=['alpha', 'beta'], figsize=(12, 3))
plt.show()
```

### **2. Random data from Normal distribution**

\begin{align} 
Y_1, ..., Y_n | \mu, \sigma^2 \sim \ \ iid \ \  Normal(\mu, \sigma^2 = 1/\tau^2)
\end{align}


#### 2a. Standart t-distribution two-sided 95% equal-tailed quantile 



MLE is given by:

\begin{align} 
\hat{\sigma}^2 = \frac{1}{n}\sum_{i=1}^{n} (y_i - \bar{y})^2 = \frac{n-1}{n}s^2
\end{align}



**Jeffreys' prior**

\begin{align} 
\pi(\mu, \sigma^2) \propto \frac{1}{(\sigma^2)^{(3/2)}} \  \ \ \ \ (\sigma^2 > 0)
\end{align}

The 95% credible interval for $\mu$


\begin{align} 
\bar{y} \pm t_{0.025,n} \frac{\hat{\sigma^2}}{ n}
\end{align}


In terms of $n, \ y,\ s^2$ The 95% credible interval for $\mu$ is given by,

\begin{align} 
\hat{\sigma^2} &= \frac{n-1}{n}s^2 \\
\frac{\sqrt{\hat{\sigma^2}}}{\sqrt{n}} &= \frac{\sqrt{\frac{n-1}{n}s^2}}{\sqrt{n}} \\
\frac{\hat{\sigma}}{\sqrt{n}} &= \frac{\sqrt{(n-1)s^2}}{n}
\end{align}

Thus, 

\begin{align} 
\bar{y} \pm t_{0.025,n} \frac{\sqrt{(n-1)s^2}}{n}
\end{align}



**Standard noninformative prior**

\begin{align} 
\pi(\mu, \sigma^2) \propto \frac{1}{\sigma^2} \  \ \ \ \ (\sigma^2 > 0) 
\end{align}


The 95% credible interval for $\mu$

\begin{align} 
\bar{y} \pm t_{0.025,n-1} \frac{s}{\sqrt{n}}
\end{align}


#### 2b. Half-width of the 95% equal-tailed posterior credible interval for $\mu$


```python
ybar = 0
s2 = 1
s = np.sqrt(s2)
qtl = 0.025
```

```python
df = np.arange(2, 11)
```

```python
ci_mu = []
sni_mu = []
for n in df:
    ci_mu.append(ybar + ss.t.ppf(1-qtl, n) * np.sqrt((n - 1) * s2) / n)
    sni_mu.append(ybar + ss.t.ppf(1-qtl, n-1) * s / np.sqrt(n))
    
fig, ax = plt.subplots()
ax.plot(df, ci_mu, label="Jeffreys' prior")
ax.plot(df, sni_mu, label="Standard non-informative")
ax.set_xlabel('n-degrees of freedom')
ax.set_ylabel(r'half-width 95% eqaul-tailed PCI for $\mu$')
ax.legend()
plt.show()
```

### **3. Graduate section only**

```python
# reading airliner dataset
df = pd.read_csv('../data/airlinerdata.txt', skipinitialspace=True, header=None)
df = (df.assign(y = df[0].apply(lambda x: x.split(' ')[0])).assign(N = df[0].apply(lambda x: x.split(' ')[3]))).drop([0],axis=1).drop([0],axis=0)
```

```python
# setting up initial data
data = {'y': df['y'].values, 'N': df['N'].values}
```

```python
# setting up hyperparameters initial values for 3 chains
inits = [{'alpha': 0.5, 'beta': 1.5},
         {'alpha': 5, 'beta': 15},
         {'alpha': 10, 'beta': 0.15}]
```

```python
## jags model string equivalent to .bug file with lambda_new
jags_model_string = '''
model {
  for (i in 1:length(y)) {
    lambda[i] ~ dgamma(alpha, beta)
    y[i] ~ dpois(N[i] * lambda[i])
  }
  lambda_new ~ dgamma(alpha, beta)
  alpha ~ dgamma(0.001,0.001)
  beta ~ dgamma(0.001,0.001)
}
'''
```

```python
# jags model initialization using pyjags
jags_model \
    = pj.Model(code=jags_model_string,
               init=inits,
               data=data,
               chains=3,
               adapt=0,
               )
```

```python
samples3 = jags_model.sample(iterations=1000000, vars=['lambda_new'])
```

```python
idata3 = az.from_pyjags(samples3)
```

```python
az.plot_trace(idata3, var_names=[ 'lambda_new'])
plt.tight_layout()
plt.show()
```

```python
## burning 10000 iterations
idata3 = idata3.sel(draw=slice(10000, 1000000))
```

```python
az.summary(idata3.posterior[['lambda_new']], stat_funcs=func_dict)
```

The approximate values for the posterior are:
* Mean:  1.247
* Standard deviation: 0.875
